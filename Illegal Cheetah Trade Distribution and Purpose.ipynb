The Distribution and Purpose of Global Illegal Cheetah Trade


Between 2010-2019, there were 4184 live cheetahs and cheetah parts traded in 56 countries in Africa, the Middle East, Asia, Europe, Oceania and North America. Most illegal cheetah trade incidents happened in Middle East, followed by Afria, Europe, Asia,Americas, and Oceania. Saudia Arabia illegally traded the largest number of Cheetahs (955) between 2010 and 2019.
Most of the Cheetahs are alive, while a large number of incidents involves Cheetah skins, bones and teeth.
